# 7. Operasyonlar

Bu bolumde takim puanlarini hesaplamak uzere veri uzerinde bazi manipulasyonlar yapacagiz.

## 7.a

- Tablodan sirasiyla yil, ev sahibi takim, ev sahibi ulke, ev sahibi golleri ve konuk takim golleri kolonlarini secip kopyalayarak yeni bir tablo olusturunuz.
- Tablonun kolonlarini `["Yil", "Takim", "Ulke", "Atilan", "Yenen"]` seklinde degistiriniz.

````python
puan_durumu = df[["Year","Home Team","Home Team Ctry","Home Goal Count","Away Goal Count"]]
````

````python
puan_durumu.columns = ["Yil", "Ev Takim", "Ev Ulke", "Ev Atilan", "Ev Yenen"]
````

````python
puan_durumu.loc[:,'Puan'] = 0
````

## 7.b

- Tablodan sirasiyla yil, konuk takim, konuk ulke, konuk takim golleri ve ev sahibi golleri kolonlarini secerek yeni bir tablo olusturunuz.
- Tablonun kolonlarini `["Yil", "Takim", "Ulke", "Atilan", "Yenen"]` seklinde degistiriniz.

````python
puan_durumu2 = df[["Year","Away Team","Away Team Ctry","Away Goal Count","Home Goal Count"]]
````

````python
puan_durumu2.columns = ["Yil2", "Takim", "Ulke", "Atilan", "Yenen"]
````

````python
puan_durumu2.loc[:,"Puan2"]=0
````

## 7.c

Girdi olarak semasi `["Yil", "Takim", "Ulke", "Atilan", "Yenen"]` seklinde bir tablo alip, tabloya `Puan` kolonunu ekleyecek bir fonksiyon yaziniz.

Puan hesaplamasi icin, Turkiye Super Liginde kullanilan konvansiyonu takip ediniz:
- Galibiyet 3 puan,
- Beraberlik 1 puan,
- Malubiyet 0 puan.

````python
def puanla(tablo):
    for i in range(0,6553):
        if puan_durumu.iloc[i,3] > puan_durumu.iloc[i,4]:
            puan_durumu.iloc[i,5] = 3
        elif puan_durumu.iloc[i,3] == puan_durumu.iloc[i,4]:
            puan_durumu.iloc[i,5] = 1
        else:
            puan_durumu.iloc[i,5] = 0
    return tablo
       
````

````python
def puanla2(tablo):
    for i in range(0,6553):
        if puan_durumu2.iloc[i,3] > puan_durumu2.iloc[i,4]:
            puan_durumu2.iloc[i,5] = 3
        elif puan_durumu2.iloc[i,3] == puan_durumu2.iloc[i,4]:
            puan_durumu2.iloc[i,5] = 1
        else:
            puan_durumu2.iloc[i,5] = 0
    return tablo
````

````python
puan_durumu.apply(puanla)
````

````python
puan_durumu2.apply(puanla2)
````

````python
puandurumuson = pd.concat([puan_durumu,puan_durumu2],axis=1)
````

````python
puandurumuson
````

# 8. Pivot Tablosu

Satir numaralarinin `Yil`, sutunlarin ise 2 seviyeli ve sirasiyla `Ulke` ve `Takim` olacak sekilde toplam puanlarin tutuldugu bir pivot tablosu olusturunuz.

## 8.a.

Her bir kulup icin puanlarin yillara sari kumulatif toplamini hesaplayiniz.



````python
bosluksuz = puandurumuson.pivot_table(index=["Yil"],columns=["Ev Ulke","Ev Takim"],values=["Puan"],aggfunc='sum')
````

````python
bosluksuz = bosluksuz.cumsum()
````

````python
bosluksuz.loc[1955] = 0
````

````python
bosluksuz = bosluksuz.fillna(method="ffill")
````

````python
bosluksuz #kumulatif olduğundan bir önceki değer atandı.
````

## 8.b.

- Kumulatif toplam tablosundaki bosluklari uygun sekilde doldurunuz.
- Elde ettiginiz sonucu yeni bir degiskene kaydediniz.
- Bosluk doldurma stratejinizi aciklayiniz.
- Tutulan verinin mahiyetini de goz onune alarak uygun bir veri tipi seciniz ve tum degerleri bu veri tipine donusturunuz.
    - *Ipucu: `.astype( ... )`*

````python
bosluksuz= bosluksuz.astype(int)
````

````python
bosluksuz.loc[   pd.IndexSlice[1968:1982], (pd.IndexSlice["Puan"],["TUR","AZE"])     ]
````
